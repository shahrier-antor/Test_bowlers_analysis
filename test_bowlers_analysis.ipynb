# Renaming the column names appropriately

```python
df = df.rename(columns={'Mat':'Match', 
                        'Inns':'Innings',
                        'Wkts':'Wickets',
                        'BBI':'Best Bowling of Innings',
                        'BBM':'Best Bowling of Matchs',
                        'SR':'Strike Rate',
                        'Ave': 'Average',
                        'Econ':'Economy Rate',
                       }
              )

display(df.head())
```

#   Removing columns from the dataframe

```python
# method 1
df = df.drop('5', axis=1)

# method 2
df.drop('10', axis=1, inplace=True)

# line 1 and line 2 both are same

# Both column named 5 and 10 is removed.

display(df.head())
```
